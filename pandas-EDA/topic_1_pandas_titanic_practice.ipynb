**Let's create a new feature.**

```python
def age_category(age):
    """"
    < 30  --> 1
    >= 30 & < 55 --> 2
    >= 55  -->3 
    """
    
    if age < 30:
        return 1 
    elif age < 55:
        return 2
    elif age >= 55:
        return 3 

    
```

```python
age_categories = []
for age in data.Age:
    age_categories.append(age_category(age))
```

```python
data['age_category'] = data['Age'].apply(age_category)
```
